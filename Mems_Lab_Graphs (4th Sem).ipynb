```python
#import python libraries
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns
print('Setup Complete')
```

## 1) When Programmable Temperature is at **50˚C** 

```python
air1=pd.read_csv('Air (50).csv', index_col='V')
air1
```

```python
h1=pd.read_csv('H2 (50).csv',index_col='V')
h1
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(16,6))
plt.title('V-I Graph at Temperature 50˚C in Air and H2')
plt.xlabel("V(V)")
#plt.xticks(rotation=-45)
plt.ylabel('I(mA)')
sns.lineplot(data=air1['I'], label='Air', color='teal')
sns.lineplot(data=h1['I'], label='H2', color='brown')
```
